```python
import pandas as pd
```

```python
cus=pd.read_csv("/content/cus_file.csv")
sal=pd.read_csv("/content/sale.csv")
exc=pd.read_csv("/content/Excl.csv")
prod=pd.read_csv("/content/product.csv")
st=pd.read_csv("/content/store.csv")
```

```python
sal=sal.drop(sal[sal["StoreKey"]==0].index)
```

```python
#joining sales with store
sal=pd.merge(sal,st,on="StoreKey",how="inner")
```

```python
sal.rename(columns={"Country":"Store_Country",'State':'Store_State'},inplace=True)
```

```python
sal.head()
```

```python
sal.shape
```

```python
#joining with customer
sal=pd.merge(sal,cus,on="CustomerKey",how="inner")
```

```python
sal.shape
```

```python
sal.rename(columns={"Country":"Customer_Country",'State':'Customer_State'},inplace=True)
```

```python
sal.head()
```

```python
#joining with product
sal=pd.merge(sal,prod,on="ProductKey",how="inner")
```

```python
sal.shape
```

```python
sal.info()
```

```python
sal[['First_name', 'Last_name']] = sal['Name'].str.split(' ', n=1, expand=True) # Changed the second positional argument to a keyword argument n=1
```

```python
sal.columns
```

```python
# Define the desired order of columns
desired_order = ['Order Number', 'Line Item', 'Order Date', 'Delivery Date',
       'CustomerKey', 'StoreKey', 'ProductKey', 'Quantity', 'Currency Code',
       'Store_Country', 'Store_State', 'Square Meters', 'Open Date', 'Gender', 'Name','First_name', 'Last_name',
       'City', 'State_Code', 'Customer_State', 'Zip_Code', 'Customer_Country', 'Continent',
       'Birthday', 'Year', 'Month', 'Age', 'Product Name', 'Brand', 'Color',
       'Unit Cost USD', 'Unit Price USD', 'SubcategoryKey', 'Subcategory',
       'CategoryKey', 'Category']

# Reorder the DataFrame columns
sal = sal[desired_order]
```

```python
sal.drop(["Name"],axis=1,inplace=True)
```

```python
sal=sal.rename(columns={'Order Number':'Order_Number', 'Line Item':'Line_Item', 'Order Date':'Order_Date', 'Delivery Date':'Delivery_Date',
      'Currency Code':'Currency_Code',
      'Square Meters':'Square_Meters', 'Open Date':'Open_Date',
       'Product Name':'Product_Name',
       'Unit Cost USD':'Unit_Cost_USD', 'Unit Price USD':'Unit_Price_USD'})
```

```python
sal.info()
```

```python
# Convert date columns to datetime objects, handling potential errors
for col in ['Birthday','Order_Date','Delivery_Date','Open_Date']:
    try:
        sal[col] = pd.to_datetime(sal[col], format='mixed', dayfirst=True, errors='coerce')
        #The errors='coerce' argument will handle errors by setting invalid parsing to NaT (Not a Time)
    except ValueError as e:
        print(f"Error converting column {col}: {e}")
        #Print any errors encountered during conversion for diagnosis
```

```python
# Sample DataFrame (replace this with your DataFrame)
# df = pd.read_csv('your_table.csv')

# Count the number of int, float, and datetime columns
int_columns = sal.select_dtypes(include=['int']).shape[1]
float_columns = sal.select_dtypes(include=['float']).shape[1]
datetime_columns = sal.select_dtypes(include=['datetime']).shape[1]
object_columns = sal.select_dtypes(include=['object']).shape[1]

# Display the counts
print(f"Number of integer columns: {int_columns}")
print(f"Number of float columns: {float_columns}")
print(f"Number of datetime columns: {datetime_columns}")
print(f"Number of object (string) columns: {object_columns}")
```
